## swagger ui
1.Open this URL in your browser:
http://127.0.0.1:8000/docs

This is called Swagger UI (FastAPI's automatic API documentation).
It's one of FastAPI's best features.
You should see a page listing all the available API endpoints.
Its like a manual for ur backend 

2.Why does FastAPI generate this page automatically?

Because when you write something like:

@app.get("/properties")

FastAPI automatically understands:

The URL
The request type
The expected input
The output format

Then it builds this documentation for you.

That's one of the reasons FastAPI is so popular.
